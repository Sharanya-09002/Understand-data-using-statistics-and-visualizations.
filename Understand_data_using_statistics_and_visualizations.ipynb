# Summary statistics

```python
import pandas as pd


df = pd.read_csv("Titanic-Dataset.csv")

# Display basic information
print(df.info())
print()
# Descriptive statistics
print(df.describe())
print()
print("\nMedian:\n", df.median(numeric_only=True))

print()
# Check for missing values
print(df.isnull().sum())
```

# Create histograms and boxplots

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for numerical features
df.hist(bins=20, figsize=(12, 10))
plt.tight_layout()
plt.show()

# Boxplot for a specific column (e.g., 'Purchase Amount')
sns.boxplot(data=df['Fare'])
plt.title("Boxplot - Fare")
plt.show()
```

# Pairplot and Correlation Matrix

```python
# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()

# Pairplot (use subset of numeric columns to avoid clutter)
selected_columns = df.select_dtypes(include='number').columns[:5]  # Pick a few for visibility
sns.pairplot(df[selected_columns])
plt.show()
```

# Identify Patterns, Trends, or Anomalies

```python
# Average Fare by Class
df.groupby('Pclass')['Fare'].mean().plot(kind='line', marker='o')
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.grid(True)
plt.show()
```
